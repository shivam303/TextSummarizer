To use the Perplexity API, you'll need to store your API key. In Colab, you can add the key to the secrets manager under the "ðŸ”‘" icon in the left panel. Give it the name `PERPLEXITY_API_KEY`. Then, you can access it securely in your code:

```python
from google.colab import userdata
import requests

PERPLEXITY_API_KEY = userdata.get('PERPLEXITY_API_KEY')

# Set up the API endpoint and headers
API_ENDPOINT = "https://api.perplexity.ai/chat/completions"
HEADERS = {
    "Authorization": f"Bearer {PERPLEXITY_API_KEY}",
    "Content-Type": "application/json"
}
```

Now, you can define a function to make a search query to the Perplexity API:

```python
def search_perplexity(query):
    data = {
        "model": "sonar-small-online", # or "sonar-medium-online" for more advanced models
        "messages": [
            {
                "role": "user",
                "content": query
            }
        ]
    }
    try:
        response = requests.post(API_ENDPOINT, headers=HEADERS, json=data)
        response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
        result = response.json()
        if 'choices' in result and len(result['choices']) > 0:
            return result['choices'][0]['message']['content']
        else:
            return "No response content found."
    except requests.exceptions.RequestException as e:
        return f"An error occurred: {e}"

# Example search
search_query = "What is the capital of France?"
answer = search_perplexity(search_query)
print(f"Query: {search_query}")
print(f"Answer: {answer}")
```
